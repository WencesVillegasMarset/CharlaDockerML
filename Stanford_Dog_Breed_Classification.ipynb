*   imágenes CUDA driver-agnostic

## Ejemplo: Clasificacion de Imagenes de Perros segun Raza

Para este ejemplo vamos a entrenar una Red Neuronal Convolucional para clasificar razas de perros.

Este es un caso de aprendizaje supervisado, ya que tendremos para cada imagen la raza correspondiente.
![alt text](https://cdn-images-1.medium.com/max/1600/1*6mPnd6tEA4EsYD1f72hGkA.png)

Una red neuronal es un modelo matematico formado por muchas unidades llamadas "neuronas".
Cada neurona esta conectada a otras mediante enlaces que representan la transformacion de los valores de entrada a los valores de salida de la red. Así cuando un valor se propaga de una neurona a otra, mediante este enlace, este se multiplica por un valor llamado "peso" y a ese valor resultante se le aplica una funcion simple como puesde ser el seno o max(0, peso*valor) (ReLu).

![alt text](https://i.imgur.com/r0VICM7.png)


![alt text](https://upload.wikimedia.org/wikipedia/commons/thumb/4/46/Colored_neural_network.svg/296px-Colored_neural_network.svg.png)

Para cada enlace que vemos en la imagen anterior, se hace la operacion de multiplicacion, se suman los resultados de cada enlace entrante a cada neurona y se aplica la funcion que se haya elegido (non-linearity). 


Basicamente un red neuronal es composicion de funciones que forma una caja negra que es capaz de aprender a modelar transformaciones muy complicadas, mientras mas capas de neuronas mas transformaciones puede llevar a cabo.
Es una caja negra que aprende a transformar los datos que le ingresamos a las etiquetas que le presentamos como verdaderas

![alt text](https://cdn-images-1.medium.com/max/1600/1*XdCMCaHPt-pqtEibUfAnNw.png)

RANDOM_SEED refiere a la semilla que va a ser usada para todos los procesos o funciones que sean random. Lo definimos asi para que sea reproducible.


````
RANDOM_SEED = 77
IMAGE_SIZE = 224
NUM_CLASSES = 10
````

Ahora viendo la estructura de carpetas que tiene nuestroo dataset, vamos a armar una tabla con dos columnas: imagen y clase. 
Lo que haremos es recorrer los directorios y por cada imagen colocar en la columna image su ruta absoluta y en la columna class el nombre de la carpeta en la que esta contenida (raza).
De esta forma tendremos a cada imagen etiquetada por el nombre de la  carpeta que la contiene, que es la raza de perro que le corresponde. 

````
full_dataset = pd.DataFrame(columns=['image','class'])
base_path = './Images'
for directory in os.listdir(base_path):
    for filename in os.listdir(os.path.join(base_path, directory)):
        current_df = pd.DataFrame({'image': [os.path.join(base_path, directory, filename)], 'class' : [os.path.split(directory)[1]]})
        full_dataset = full_dataset.append(current_df, ignore_index=True)
        

full_dataset['class'] = full_dataset['class'].replace(regex=r"^n[0-9]+-", value='')
razas_list = ['collie','German_shepherd', 'Siberian_husky', 'Doberman', 'Great_Dane', 'Pekinese', 'beagle', 'Border_collie', 'Rhodesian_ridgeback', 'schipperke']
full_dataset = full_dataset.loc[full_dataset['class'].isin(razas_list), :]
full_dataset = shuffle(full_dataset, random_state=RANDOM_SEED)

training_partition, test_df = train_test_split(full_dataset, test_size=0.2, random_state=RANDOM_SEED, stratify=full_dataset['class'])

train_df, valid_df = train_test_split(training_partition, test_size=0.25, random_state=RANDOM_SEED, stratify=training_partition['class'])

````

Definimos algunas funciones auxiliares que van a ser utiles mas adelante.  
-  plot_sample_images : toma un csv y lee las primeras 20 filas para mostrar las imagenes y su clase como titulo.

-  plot_train_history : toma un objeto History de la libreria Keras y grafica loss y accuracy para training y validation reportadas durante el entrenamiento.
-  preprocess_image: tomamos una imagen y la dividimos por 255, haciendo que todos los pixeles tengan valor entre 0 y 1. Estoo generalmente ayuda a la convergencia de la red y prevenir ciertos comportamientos no deseados durante el entrenamiento. (http://cs231n.github.io/neural-networks-2/#datapre)

````
def plot_sample_images(image_data):
    w=10
    h=10
    fig=plt.figure(figsize=(15, 12))
    columns = 4
    rows = 5
    for i in range(1, columns*rows +1):
        img = plt.imread(image_data.iloc[i,0])
        fig.add_subplot(rows, columns, i)
        plt.imshow(img)
        plt.title(image_data.iloc[i,1])
        plt.axis('off')
    plt.show()
            
def plot_train_history(history):
    fig, ax = plt.subplots(1, 2, figsize=(15,5))
    ax[0].set_title('loss')
    ax[0].plot(history.epoch, history.history["loss"], label="Train loss")
    ax[0].plot(history.epoch, history.history["val_loss"], label="Validation loss")
    ax[1].set_title('acc')
    ax[1].plot(history.epoch, history.history["acc"], label="Train acc")
    ax[1].plot(history.epoch, history.history["val_acc"], label="Validation acc")
    ax[0].legend()
    ax[1].legend()

def preprocess_image(img):
    return np.divide(img, 255.0)
````

Ahora definimos Generadores de Imagenes que van a hacer el trabajo de cargar las imagenes en memoria, preprocesarlas y pasarlas a la red.
Para esto definimos un ImageDataGenerator para cada particion del dataset.
Usamos el metodo flow_from_dataframe para usar un DataFrame como la fuente de datos. Es decir que el generador va a leer las filas del DataFrame y va a extraer la imagen de la ruta que especifica la primer columna y le va a asignar la etiqueta correspondiente. Para eso especificamos cual va a ser la columna X (datos | input) y la columna Y (etiqueta). Tambien especificamos un batch_size, que representa la cantidad de imagenes que se le van a pasar a la red a cada paso.

![texto alternativo](https://cdn-images-1.medium.com/max/1600/1*HpvpA9pBJXKxaPCl5tKnLg.jpeg)

````
#definimos un generado base
image_generator = image.ImageDataGenerator(preprocessing_function=preprocess_image)
#definimos un generador para los datos de entrenamiento
train_generator = image_generator.flow_from_dataframe(train_df, x_col='image', y_col='class', target_size=(IMAGE_SIZE, IMAGE_SIZE),
                                                      batch_size=32, class_mode='categorical', random_seed=RANDOM_SEED, rotation_range=10,
                                                      zoom_range = 0.1, width_shift_range=0.2, height_shift_range=0.2, horizontal_flip=True)
#definimos un generador para los datos de validacion
valid_generator = image_generator.flow_from_dataframe(valid_df, x_col='image', y_col='class', target_size=(IMAGE_SIZE, IMAGE_SIZE),
                                                      batch_size=32, class_mode='categorical', random_seed=RANDOM_SEED)
#definimos un generador para los datos de testeo
test_generator = image_generator.flow_from_dataframe(test_df, x_col='image', y_col='class', target_size=(IMAGE_SIZE, IMAGE_SIZE),
                                                     batch_size=1, class_mode='categorical', random_seed=RANDOM_SEED, shuffle=False)
````

Esto hara que el generador asigne para cada raza, que se encuentra como una cadena, una etiqueta numerica (un entero). Esta asignacion se puede ver en el atributo class_indices del objeto.

Ahora llega el momento de definir nuestro modelo.
Para este caso aplicaremos la tecnica de transfer learning y cargaremos un modelo pre-entrenado en el dataset Imagenet.
![texto alternativo](https://cdn-images-1.medium.com/max/2600/1*Z11P-CjNYWBofEbmGQrptA.png)
Este modelo es MobileNet en su segunda version. MobileNet es una arquitectura de Red Convolucional que se caracteriza por ser pensada para correr en dispositivos moviles. Es por esto que incorpora bloques alternativos a las redes convolucionales convencionales que tienden a bajar significativamente el numero de parametros y la velocidad tanto de entrenamiento como de inferencia.

Primero cargamos MobileNet como extractor de features.
Y luego agregamos las capas que necesitamos especificamente para hacer clasificacion en nuestro problema.


````
#Primero cargamos MobileNet sin sus capas FC, definimos el tamaño de los inputs y que sus pesos sean preentrenados
encoder = MobileNetV2(weights = "imagenet", include_top = False, input_shape = (IMAGE_SIZE, IMAGE_SIZE, 3))

#a todas las capas que conforman este "encoder" les especificamos que no sean entrenables
for layer in encoder.layers:
    layer.trainable = False #no seran entrenables, no se calcularan gradientes para estas capas
    
#luego creamos un objeto Sequential, que sera el objeto que contendrá a la todas las capas del modelo
model = Sequential()

# agregamos como primer capa de nuestro modelo al encoder completo
model.add(encoder)

# agregamos una capa de max pooling global a aplicar al output del encoder
model.add(GlobalMaxPooling2D())

# agregamos una capa de dropout, para incluir regularizacion
model.add(Dropout(0.5))

#agregamos la ultima capa que sera la encargada de calcular las probabilidades finales, es por eso que ponemos NUM_CLASSES
model.add(Dense(NUM_CLASSES, activation='softmax'))
model.summary()
````

Aqui definimos algunas callbacks utiles para el entrenamiento de la red.
Una es reduce, que lo que hace es monitorear las metricas de entrenamiento, si esta se queda estancada, reduce el learning rate.
La segunda implementa EarlyStopping, esta monitorea las metricas y cuando cierta metrica objetivo se estanca, para el entrenamiento. Esto es util para prevenir el "sobreentrenamiento" de la red que puede causar overfitting y tambien evitar perder el tiempo entrenando algo que esta estancado en un optimo y no poder hacer nada al respecto.

````
reduce = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.1,
    patience=5,
    verbose=1, 
    mode='auto'
)
earlystop = EarlyStopping(
    monitor='val_loss',
    min_delta=0.001,
    patience=30,
    verbose=1,
    mode='auto'
)
sgd = keras.optimizers.SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
adam = keras.optimizers.Adam(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=None, decay=0.0, amsgrad=False)
````

Antes de comenzar con el entrenamiento de nuestro modelo, debemos definir nuestra funcion de costo, es decir la funcion que usaremos para calcular el error de prediccion y que diferenciaremos a cada paso ya que modela el error con respecto a cada parametro.
Tambien definiremos un optimizador, este optimizador es basicamente la formula que nosotros usaremos para penalizar a los parametros una vez calculada la influencia que estos tuvieron sobre el error (gradiente). Existen muchos optimizadores, algunos muy simples y otros con formulas mas complejas. 
Adam es un optimizador muy usado ya que ha probado ser muy efectivo en la mayoria de casos de uso y requiere poco tuning.
Tambien usaremos una metrica con la cual evaluar el rendimiento de nuestra red, en este caso usamos la convencional accuracy. 

````
# for layer in model.layers:
#     layer.trainable = True
model.compile(loss='categorical_crossentropy', optimizer=adam, metrics=['accuracy'])
````

Ahora estamos listos para entrenar nuestra red. 
Para ello especificamos el generador de entrenamiento, el generador de validacion, las callbacks que vamos a usar y los epochs por los que vamos a entrenar.

````
history = model.fit_generator(train_generator, validation_data=valid_generator, callbacks=[earlystop], validation_steps=100, steps_per_epoch=1000, epochs=15)
````

Definimos una funcion para cargar el modelo y una funcion para preprocesar las imagenes entrantes

````
def load_model():
	# cargamos el modelo de keras que tenemos en el repo poniendo su ruta como argumento en leado model
	#esto solo se ejecuta una vez ya que no necesitamos cargar el modelo con cada peticion
	global model
	model = keras.models.load_model('dog_classifier.h5')

def prepare_image(image, target_size):
	# si el modo de la imagen no es RGB la convertimos a ese tipo
	if image.mode != "RGB":
		image = image.convert("RGB")
	# ahora le hacemos el preprocessing necesario para que entre a la red.
	image = image.resize(target_size)
	image = keras.preprocessing.image.img_to_array(image)
	image = np.expand_dims(image, axis=0)
	image = image/255.0

	return image
````

Ahora definimos una ruta para aceptar peticiones de tipo POST con imagenes para clasificar

````
@app.route("/predict", methods=["POST"]) #decorador flask
def predict():
	# inicializamos el diccionario que va a devolver el metodo
	data = {"success": False}

	if flask.request.method == "POST":
		#nos aseguramos de que haya llegado la imagen por POST
		if flask.request.files.get("image"):
			# leemos la imagen y la transformamos a formato PIL
			image = flask.request.files["image"].read()
			image = Image.open(io.BytesIO(image))

			# preprocesamos la imagen
			image = prepare_image(image, target_size=(224, 224))

			# clasificamos la imagen
            preds = model.predict(image)
			
			# procesamos los resultados de predciccion para que sean legibles
			full_result, class_predicted = decode_predictions(preds)
			# armamos el diccionario que vamos a devolver al usuario
			data["predictions"] = class_predicted
			data['probabilities'] = full_result
			data["success"] = True

	# returnamos los resultados en formato json
	return flask.jsonify(data)
````

Ahora definiremos lo que se ejecutará cuando se llame a la script de la siguiente manera

